# preprocessing data

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler
y= df["target"]   #depedent variable 
x = df.drop('target',axis=1)   #independent variable 
xtr,xtst,ytr,ytst = tts(x, y, test_size=0.15)    #splitting data 85% vs 15%
scaler=StandardScaler()      #data standardization 
xtr=scaler.fit_transform(xtr)
xtst=scaler.fit_transform(xtst)
```

# SVM

```python
from sklearn.svm import SVC   #support vector classifier 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
svc =  SVC()
svc.fit(xtr, ytr)
print('training score',svc.score(xtr,ytr))
yhat= svc.predict(xtst)
print("confussion matrix")
print(confusion_matrix(ytst, yhat))    #ytst is actual result and yhat is predicted result 
print("-------------------------------------------")
print("Accuracy of Support Vector Classifier:",accuracy_score(ytst, yhat))
print("-------------------------------------------")
print(classification_report(ytst,yhat))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
tr=DecisionTreeClassifier()
tr.fit(xtr,ytr)
print('training score',tr.score(xtr,ytr))
yhat=tr.predict(xtst)
```

```python
print(classification_report(yhat,ytst))
print('##################')
print(confusion_matrix(yhat,ytst))
print('accuracy score',accuracy_score(yhat,ytst))
```

# logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr=LogisticRegression()
lr.fit(xtr,ytr)
print('training',lr.score(xtr,ytr))
yhat=lr.predict(xtst)
print('testing',lr.score(xtst,ytst))
```

```python
print(classification_report(yhat,ytst))
print(confusion_matrix(yhat,ytst))
print('accuracy score',accuracy_score(yhat,ytst))
```
